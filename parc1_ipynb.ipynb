```
import copy
import matplotlib
import matplotlib.pyplot as plt
from scipy import linalg #行列計算用
import numpy as np

N = 32
h = 0.1
v = np.array([min(x, N-x) for x in range(N)], dtype='float64')
A = np.zeros((N, N))

# 行列Aを作る
for i in range(N):
    i1 = (i + 1) % N
    i2 = (i - 1 + N) % N
    #print('~~~~~~~~~~')
    #print(A)
    A[i][i] = 1.0 - 2.0*h
    #print(i,i,'代入',A)
    A[i][i1] = h
    #print(i,i1,'代入',A)
    A[i][i2] = h
    #print(i,i2,'代入',A)

r = []
#for i in range(10):
for i in range(1000):
    v = A.dot(v) # Aをかけると次のステップの状態が得られる
    if (i % 100) == 0:
        r.append(copy.copy(v))

for s in r:
    plt.plot(s)


w, v = linalg.eigh(A, eigvals=(N-1,N-1))
print(w) #=> [1.]　#行列Aの最大固有値
print(v) #対応する固有ベクトル=定常状態
w, v = linalg.eigh(A, eigvals=(N-2,N-2))
print(w) #=> [1.] 
print(v)
```

```
#熱伝導方程式
import copy
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#v(温度)とhを計算する関数を定義
def calc(v, h):
    v2 = copy.copy(v)
    print('v2更新')
    print('v2=',v2)
    N = len(v)
    for i in range(N): #iは空間の位置番号
        print('~~~空間伝搬ループ 空間位置=',i,'~~~')
        i1 = (i+1) % N #(i+1)をNで割った余り 1スタートで0~31を繰り返す
        i2 = (i - 1 + N) % N #31スタートで0~31~0を繰り返す
        #print([i2,i,i1])
        print('代入前v[',i,']=',v[i])
        v[i] = v2[i] + (v2[i1] - 2.0*v2[i] + v2[i2])*h #v2のi2,i,i1からv[i]を計算する（iが0~31まで回る（熱が空間伝搬する）度に更新される）
        #print('i=',i)
        #print('i1=',i1)
        #print('i2=',i2)
        #print('v2[',i,']=',v2[i])
        #print('v2[',i1,']=',v2[i1])
        #print('v2[',i2,']=',v2[i2])
        #print('v1更新')
        print('代入後v[',i,']=',v[i])
        #print('v=',v)
###########################################

N = 32 #空間を32分割
v = np.array([min(x, N-x) for x in range(N)], dtype='float64')
#testv = np.array([x for x in range(N)], dtype='float64')
print('v=',v)
#print('testv=',testv)

h = 0.1 #時間刻み幅
r = []

for i in range(3): #1000回繰り返す
#for i in range(1000): #1000回繰り返す
    print('~~~時間発展ステップループ 時間=',i,'~~~')
    calc(v, h)
    #if (i % 100) == 0:
    if (i % 1) == 0:
        r.append(copy.copy(v))

for s in r:
    plt.plot(s)
```

```
#####class勉強#####
class TestClass: #頭大文字が慣習
    pass
x = TestClass() #変数xにTestClassのインスタンス代入

#classの属性
class Character: #classを定義する
    pass

chara1 = Character() #chara1変数にCharacterインスタンスを作成
chara1.name = 'Kazuma' #chara1インスタンスのname属性の値を’Kazuma’に設定
chara1.age = 16 #chara1インスタンスのage属性の値を16に設定
chara1.sex = 'male' #chara1インスタンスのsex属性の値を’male’に設定
print(chara1.name)

chara2 = Character() #chara2変数にCharacterインスタンスを作成
chara2.name = 'Megumi' #chara1とは別の値を設定できる
chara2.age = 13
chara2.sex = 'female'
print(chara2.name)

#classのメソッド
class Character:
    def set_data(self, set_name, set_age, set_sex): #属性に値を一括で設定するメソッドを定義 #第1引数に「self」(メソッドを呼び出したインスタンス自身)というオブジェクトを指定しておく
        self.name = set_name #selfには実際にはインスタンスが入る。
        self.age = set_age #属性の説明時に各属性に値を設定したのと同じこと。
        self.sex = set_sex #メソッドの引数に自分自身を入れることで、自分自身に設定されている属性の値を利用できる
    def print_data(self):
        print("私は{}。{}歳。{}です。".format(self.name, self.age, self.sex)) # 各属性の値をprint。

chara3 = Character() #新しいインスタンスを作成
chara3.set_data('Kazuma', 16, 'male') #一括で属性に値を設定　１つ目のメソッド実行！
chara3.print_data() #2つ目のメソッド実行！

#classのメソッド2
class TestClass:
    x = "変数1"
    def test_method1(self): #クラス内の関数をメソッド呼ぶ
        print(self.x)
    def test_method2(self, arg1):
        print("引数1:" + arg1)
testClass = TestClass()
testClass.test_method1() # ⇒ 変数1
testClass.test_method2("引数Test")

#classのinit
class Character:
    def __init__(self, name, age, sex): #コンストラクタ(インスタンス作成時に自動で呼び出されるメソッド)でインスタンスを初期化
        self.name = name
        self.age = age
        self.sex = sex 
    def print_data(self):
        print("私は{}。{}歳。{}です。".format(self.name, self.age, self.sex))

chara4 = Character('Kazuma', 16, 'male') # インスタンス作成時にコンストラクタが呼び出される 
chara4.print_data()

#classの継承
class Villager(Character): #Characterクラスを継承
    def __init__(self, name, age, sex, line): #コンストラクタを拡張(line追加)
        super().__init__(name, age, sex) #super()で継承元のクラスを呼び出せる
        self.line = line #台詞を設定
    def speak(self): #台詞を話すメソッドを定義
        print("{}「{}」".format(self.name, self.line))

villager1 = Villager("apple seller", 35, "male", "おう！リンゴでもどうだい？") #インスタンス作成時、属性に値を設定 
villager1.speak() #台詞を話すメソッド

villager1.print_data()

#classの継承2
class Hiro(Character): #Characterクラスを継承
    def __init__(self, name, age, sex, hp, ex, mp): #コンストラクタを拡張
        super().__init__(name, age, sex) #super()で継承元のクラスを呼び出せる
        self.hp = hp #体力を設定
        self.ex = ex #経験値を設定
        self.mp = mp #魔力を設定
    def heal(self): #魔力に応じて体力を回復するメソッド
        self.hp += self.mp # 体力を魔力分回復
        print("HPが{}回復".format(self.mp)) #回復量を表示
        print("現在のHP：{}".format(self.hp)) #回復後の体力を表示

hiro1 = Hiro('Kazuma', 16, 'male', 100, 200, 30) 
hiro1.heal() #魔力分の体力を回復するメソッド

#復習 torchvision.transforms
class Plus2(object):
    def __init__(self):
        pass
    def __call__(self, x):
        data = x + 2
        return data

trans = Plus2()
x = 9
data = trans(x)
print(x)
print(data)


#class MyClass:
#    """A simple example class"""         # 三重クォートによるコメント
#    def __init__(self):                  # コンストラクタ
#        self.name = ""
#    def getName(self):                   # getName()メソッド
#        return self.name
#    def setName(self, name):             # setName()メソッド
#        self.name = name
#a = MyClass()                            # クラスのインスタンスを生成
#a.setName("Tanaka")                      # setName()メソッドをコール
#print a.getName()                        # getName()メソッドをコール
```
